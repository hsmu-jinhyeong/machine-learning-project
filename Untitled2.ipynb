```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
import sklearn
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df_b = pd.read_csv("/content/drive/MyDrive/tft_match_data/tft_match_databronze_match_ids_match_ids.csv")
df_s = pd.read_csv("/content/drive/MyDrive/tft_match_data/tft_match_datasilver_match_ids_match_ids.csv")
df_g = pd.read_csv("/content/drive/MyDrive/tft_match_data/tft_match_datagold_match_ids_match_ids.csv")
df_p = pd.read_csv("/content/drive/MyDrive/tft_match_data/tft_match_dataplatinum_match_ids_match_ids.csv")
df_d = pd.read_csv("/content/drive/MyDrive/tft_match_data/tft_match_datadiamond_match_ids_match_ids.csv")

all_dfs = [df_b, df_s, df_g, df_p, df_d]

# 랭크 게임 이외의 게임에 대한 데이터 삭제
for dataframe in all_dfs:
    dataframe.drop(dataframe[dataframe['tft_set_core_name'] == 'TFTSet10'].index, inplace=True)
    dataframe.drop(dataframe[dataframe['tft_game_type'] == 'pairs'].index, inplace=True)
    dataframe.drop(dataframe[dataframe['tft_game_type'] == 'turbo'].index, inplace=True)
    dataframe.drop(dataframe[dataframe['tft_game_type'] == 'pve'].index, inplace=True)

# 필요한 데이터만 추출하여 새로운 DataFrame 생성
df_B = df_b[['placement', 'units', 'traits']].copy()
df_S = df_s[['placement', 'units', 'traits']].copy()
df_G = df_g[['placement', 'units', 'traits']].copy()
df_P = df_p[['placement', 'units', 'traits']].copy()
df_D = df_d[['placement', 'units', 'traits']].copy()
```

```python
# JSON 문자열 컬럼을 파이썬 객체로 파싱하는 함수
def parse_json_columns(df, df_name="데이터프레임"):
    print(f"\n--- '{df_name}' 데이터프레임 'units'와 'traits' 컬럼 JSON 파싱 시작 ---")
    for col in ['units', 'traits']:
        if col in df.columns:
            df[col] = df[col].astype(str).apply(
                lambda x: json.loads(x) if x.strip() and x.strip().startswith('[') else []
            )
            print(f"'{df_name}'의 '{col}' 컬럼 파싱 완료. 첫 번째 셀 타입: {type(df[col].iloc[0])}")
        else:
            print(f"경고: '{df_name}'에 '{col}' 컬럼이 없습니다. 데이터셋을 확인해주세요.")
    print(f"--- '{df_name}' 데이터프레임 'units'와 'traits' 컬럼 JSON 파싱 완료 ---")
    return df

# --- 1. 데이터 전처리 및 특징 추출 함수 ---
def preprocess_and_extract_features(df):
    print(f"\n--- 데이터 전처리 및 특징 추출 시작 (DataFrame Shape: {df.shape}) ---")

    if not df.empty:
        first_row_units = df.iloc[0]['units'] if 'units' in df.columns else "컬럼 없음"
        first_row_traits = df.iloc[0]['traits'] if 'traits' in df.columns else "컬럼 없음"
        print(f"디버깅: 첫 번째 행 'units' 컬럼 (타입: {type(first_row_units)}): {str(first_row_units)[:100]}...")
        print(f"디버깅: 첫 번째 행 'traits' 컬럼 (타입: {type(first_row_traits)}): {str(first_row_traits)[:100]}...")

    all_champion_ids = set()
    all_trait_names = set()

    for index, row in df.iterrows():
        if 'units' in row and isinstance(row['units'], list):
            for unit in row['units']:
                if 'character_id' in unit and unit['character_id'] and isinstance(unit['character_id'], str):
                    all_champion_ids.add(unit['character_id'])
        if 'traits' in row and isinstance(row['traits'], list):
            for trait in row['traits']:
                if 'name' in trait and trait['name'] and isinstance(trait['name'], str):
                    all_trait_names.add(trait['name'])

    print(f"디버깅: 발견된 유니크 챔피언 ID 개수: {len(all_champion_ids)}")
    print(f"디버깅: 발견된 유니크 특성 이름 개수: {len(all_trait_names)}")

    if not all_champion_ids:
        print("경고: 데이터에 유효한 챔피언 정보가 부족합니다. 클러스터링 불가.")
        return None, None, None
    if not all_trait_names:
        print("경고: 데이터에 유효한 특성 정보가 부족합니다. 클러스터링 불가.")
        return None, None, None

    processed_decks_features = []
    for index, row in df.iterrows():
        deck_features = {}
        for champ_id in all_champion_ids:
            deck_features[f'unit_{champ_id}'] = 0
        if 'units' in row and isinstance(row['units'], list):
            for unit in row['units']:
                if 'character_id' in unit and 'tier' in unit and isinstance(unit['tier'], (int, float)):
                    if unit['character_id'] in all_champion_ids:
                        deck_features[f'unit_{unit["character_id"]}'] = unit['tier']

        for trait_name in all_trait_names:
            deck_features[f'trait_{trait_name}'] = 0
        if 'traits' in row and isinstance(row['traits'], list):
            for trait in row['traits']:
                if 'name' in trait and 'tier_current' in trait and isinstance(trait['tier_current'], (int, float)):
                    if trait['name'] in all_trait_names:
                        deck_features[f'trait_{trait["name"]}'] = trait['tier_current']

        processed_decks_features.append(deck_features)

    if not processed_decks_features:
        print("오류: 특징 데이터가 생성되지 않았습니다. 클러스터링 불가.")
        return None, None, None

    features_df = pd.DataFrame(processed_decks_features)

    object_cols = features_df.select_dtypes(include='object').columns
    if len(object_cols) > 0:
        print(f"경고: 숫자형이 아닌 컬럼 발견 ({len(object_cols)}개). 숫자로 변환 시도.")
        for col in object_cols:
            features_df[col] = pd.to_numeric(features_df[col], errors='coerce').fillna(0)

    if not np.issubdtype(features_df.values.dtype, np.number):
        print("치명적 오류: 모든 변환 시도 후에도 숫자형이 아닌 데이터가 존재합니다.")
        return None, None, None

    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(features_df)
    print("데이터 스케일링 완료.")
    return features_df, scaled_features, scaler

# --- 2. 엘보우 그래프 생성 함수 ---
def plot_elbow_method(scaled_features, max_k_range=10):
    print("\n--- 엘보우 메서드를 이용한 최적 K 값 탐색 ---")
    inertia = []
    max_k_actual = min(max_k_range, scaled_features.shape[0] - 1)

    if max_k_actual < 1:
        print("경고: 샘플 개수가 너무 적어 엘보우 메서드를 실행할 수 없습니다 (최소 2개 샘플 필요).")
        return

    for k in range(1, max_k_actual + 1):
        try:
            kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
            kmeans.fit(scaled_features)
            inertia.append(kmeans.inertia_)
        except ValueError as e:
            print(f"오류 발생 (K={k}): {e}. 이 K 값에서는 클러스터링 불가.")
            break

    if not inertia:
        print("엘보우 그래프를 그릴 데이터가 없습니다.")
        return

    plt.figure(figsize=(10, 6))
    plt.plot(range(1, len(inertia) + 1), inertia, marker='o')
    plt.xlabel('클러스터 개수 (K)')
    plt.ylabel('응집도 (Inertia)')
    plt.title('최적 K를 위한 엘보우 메서드')
    plt.xticks(range(1, len(inertia) + 1))
    plt.grid(True)
    plt.show()
    print("엘보우 그래프가 생성되었습니다. 그래프를 보고 최적의 K 값을 직접 선택해주세요.")


# --- 3. K-Means 클러스터링 및 결과 분석 함수 ---
def perform_kmeans_and_analyze(df, features_df, scaled_features, scaler, optimal_k, df_name="데이터셋"):
    print(f"\n--- {df_name} K-Means 클러스터링 및 결과 분석 시작 (K={optimal_k}) ---")

    if scaled_features is None or scaled_features.shape[0] < optimal_k or optimal_k <= 0:
        print(f"오류: {df_name} 클러스터링을 위한 데이터가 부족하거나 (샘플 수: {scaled_features.shape[0]}, 요구 K: {optimal_k}) K 값이 유효하지 않습니다.")
        return None

    kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init='auto')
    clusters = kmeans.fit_predict(scaled_features)
    df.loc[:,'cluster'] = clusters
    print(f"{df_name} K-Means 클러스터링 완료. 클러스터 라벨이 DataFrame에 추가되었습니다.")

    print("\n" + "="*50)
    print(f"             {df_name} TFT 덱 클러스터링 분석 결과")
    print("="*50)

    # 덱 분포 및 평균 순위
    print("\n### 1. 클러스터별 덱 분포 및 평균 순위")
    cluster_summary = df.groupby('cluster')['placement'].agg(['count', 'mean']).reset_index()
    cluster_summary.columns = ['클러스터', '덱 개수', '평균 순위']
    cluster_summary['평균 순위'] = cluster_summary['평균 순위'].round(2)
    cluster_summary = cluster_summary.sort_values(by='평균 순위').reset_index(drop=True) # 순위 기준 정렬
    print(cluster_summary.to_string(index=False))
    print("\n* 평균 순위는 낮을수록 좋은 성능을 의미합니다.")

    # 클러스터별 핵심 특징 요약
    cluster_centers_original_scale = scaler.inverse_transform(kmeans.cluster_centers_)
    cluster_analysis_df = pd.DataFrame(cluster_centers_original_scale, columns=features_df.columns)

    print("\n\n### 2. 각 클러스터의 주요 특징")
    print("각 클러스터를 대표하는 챔피언 (평균 별 등급) 및 특성 (평균 활성화 레벨)입니다.")

    num_top_features_display = 8 # 출력할 상위 챔피언/특성 개수

    # 클러스터별 주요 챔피언을 저장할 리스트
    all_cluster_top_champs_data = []

    # cluster_summary의 클러스터 순서대로 반복하여 평균 순위 정보를 가져옴
    for i in cluster_summary['클러스터']:
        # 해당 클러스터의 평균 순위 가져오기
        avg_placement = cluster_summary[cluster_summary['클러스터'] == i]['평균 순위'].iloc[0]

        print(f"\n--- 클러스터 {i} (덱 개수: {cluster_summary[cluster_summary['클러스터'] == i]['덱 개수'].iloc[0]}개, 평균 순위: {avg_placement}위) ---")

        champ_features = cluster_analysis_df.loc[i].filter(like='unit_').sort_values(ascending=False)
        print("  **주요 챔피언 (평균 별 등급):**")
        top_champs_count = 0

        # 클러스터 번호와 평균 순위를 딕셔너리에 먼저 추가
        current_cluster_champs_dict = {'클러스터': i, '평균 순위': avg_placement}

        for feature_name, value in champ_features.items():
            if value > 0.05:
                clean_name = feature_name.replace('unit_TFT14_', '').replace('unit_', '')
                print(f"    - {clean_name}: {value:.2f}")
                current_cluster_champs_dict[f'챔피언 {top_champs_count + 1}'] = f'{clean_name} ({value:.2f})'
                top_champs_count += 1
            if top_champs_count >= num_top_features_display:
                break

        while top_champs_count < num_top_features_display:
            current_cluster_champs_dict[f'챔피언 {top_champs_count + 1}'] = np.nan
            top_champs_count += 1

        all_cluster_top_champs_data.append(current_cluster_champs_dict)


        trait_features = cluster_analysis_df.loc[i].filter(like='trait_').sort_values(ascending=False)
        print("  **주요 특성 (평균 활성화 레벨):**")
        top_traits_printed = 0
        for feature_name, value in trait_features.items():
            if value > 0.05:
                clean_name = feature_name.replace('trait_TFT14_', '').replace('trait_', '')
                print(f"    - {clean_name}: {value:.2f}")
                top_traits_printed += 1
            if top_traits_printed >= num_top_features_display:
                break
        if top_traits_printed == 0:
            print("    - 해당 클러스터에 두드러지는 특성이 없습니다.")

    # --- 클러스터별 주요 챔피언 데이터프레임 생성 및 출력 ---
    print("\n\n### 3. 클러스터별 주요 챔피언 요약 (DataFrame)")
    cluster_top_champs_df = pd.DataFrame(all_cluster_top_champs_data)

    # '클러스터', '평균 순위' 컬럼을 가장 앞으로 이동
    cols_order = ['클러스터', '평균 순위'] + [col for col in cluster_top_champs_df.columns if col not in ['클러스터', '평균 순위']]
    cluster_top_champs_df = cluster_top_champs_df[cols_order]

    print(cluster_top_champs_df.to_string(index=False))
    print("\n* 챔피언 옆 괄호 안의 숫자는 해당 챔피언의 평균 별 등급을 의미합니다.")
    print("\n* 평균 순위는 낮을수록 좋은 성능을 의미합니다.")
    print("\n* 아직 클러스터 이름이 부여되지 않았습니다. 분석 결과를 보고 직접 이름을 지정해주세요.")


    print("\n" + "="*50)
    print("분석 요약: 각 클러스터의 주요 챔피언과 특성을 통해 덱의 핵심 시너지를 파악할 수 있습니다.")
    print("평균 순위가 낮은 클러스터는 현재 메타에서 강력한 덱 유형일 수 있습니다.")
    print("="*50)

    return cluster_top_champs_df
```

```python
df_B = parse_json_columns(df_B, "df_B (브론즈)")
df_S = parse_json_columns(df_S, "df_S (실버)")
df_G = parse_json_columns(df_G, "df_G (골드)")
df_P = parse_json_columns(df_P, "df_P (플래티넘)")
df_D = parse_json_columns(df_D, "df_D (다이아몬드)")
```
